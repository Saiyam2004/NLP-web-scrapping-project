```python

!pip install requests beautifulsoup4 pandas lxml

import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://www.espncricinfo.com/rss/content/story/feeds/0.xml"
response = requests.get(url)


soup = BeautifulSoup(response.content, "xml")


items = soup.find_all("item")


data = []
for item in items[:10]:
    title = item.title.text
    description = item.description.text


    if " - " in title:
        match, score = title.split(" - ", 1)
    else:
        match, score = title, description

    data.append([match.strip(), score.strip()])


df = pd.DataFrame(data, columns=["Match", "Score/Update"])


import IPython.display as display
display.display(df)
```

```python

!pip install requests beautifulsoup4 pandas lxml

import requests
from bs4 import BeautifulSoup
import pandas as pd


url = "https://books.toscrape.com/catalogue/page-1.html"


def rating_to_num(rating_class):
    ratings = {"One": 1, "Two": 2, "Three": 3, "Four": 4, "Five": 5}
    for key, val in ratings.items():
        if key in rating_class:
            return val
    return 0


books = []
page = 1
while True:
    res = requests.get(f"https://books.toscrape.com/catalogue/page-{page}.html")
    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.content, "html.parser")
    articles = soup.find_all("article", class_="product_pod")

    if not articles:
        break

    for article in articles:

        title = article.h3.a["title"]


        price = article.find("p", class_="price_color").text
        price_value = float(price.replace("Â£", "").strip())


        rating_class = article.p["class"]  # ['star-rating', 'Three']
        rating = rating_to_num(" ".join(rating_class))

        books.append([title, price_value, rating])

    page += 1


df = pd.DataFrame(books, columns=["Title", "Price (Â£)", "Rating"])
df.to_csv("books.csv", index=False)


avg_price = df["Price (Â£)"].mean()
best_book = df.loc[df["Rating"].idxmax()]

print(f"ðŸ“š Average Price: Â£{avg_price:.2f}")
print("\nðŸŒŸ Best Rated Book:")
print(best_book)


df.head()
```
